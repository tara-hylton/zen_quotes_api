```python
import pandas as pd
import requests
```

Connecting to zen quotes api and putting into a dataframe using pandas

```python
url = "https://zenquotes.io/api/quotes"


response = requests.get(url)
    
if response.status_code == 200:
    quote_data = response.json()

else:
    print(f"Failed to retrieve data {response.status_code}")

quote_df = pd.DataFrame(quote_data)
print(quote_df)
```

Connecting to zen quotes api and using a function to pull a random quote

```python
url_random = "https://zenquotes.io/api/random"
def get_random_quote(url_random):
    
    response = requests.get(url_random)
    if response.status_code == 200:
        random_quote = response.json()
        return random_quote
    elif response.status_code == 429:
        print("Too many requests, please wait 30 seconds")
        
    else:
        print(f"Failed to retrieve data {response.status_code}")

print(get_random_quote(url_random))
```
